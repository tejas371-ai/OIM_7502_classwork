```python
df.columns = df.columns.str.strip()    # Remove any leading/trailing spaces in column names
df.columns
```

```python
df[['Taxes', 'Sell']].dtypes
```

```python
df['Taxes'] = pd.to_numeric(df['Taxes'], errors='coerce')
df['Sell'] = pd.to_numeric(df['Sell'], errors='coerce')
```

```python
df_model = df[['Taxes', 'Sell']].dropna()  #Drop NaN values
```

```python
X = df_model[['Taxes']]        # independent variable
y = df_model['Sell']           # dependent variable
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

### Result:
The p-value is low (typically < 0.05), meaning Taxes is a statistically significant predictor of selling price.
The coefficient is positive, which means as property taxes increase, selling price tends to increase as well.
The R-squared value is moderate (around 0.5 in many cases), meaning taxes explain a fair portion of the variation in selling price but not all of it.

### Conclusion:
Taxes is a good predictor of Sell price, but not the only factor.
Other variables (like living area, bedrooms, age, etc.) may improve the prediction accuracy when included in a multiple regression model.

```python
print(df[['Age', 'Sell']].isna().sum())
print(np.isinf(df[['Age', 'Sell']]).sum())

# Drop rows with NaN or inf in Age or Sell
df_age_model = df[['Age', 'Sell']].replace([np.inf, -np.inf], np.nan).dropna()

X = df_age_model[['Age']]
y = df_age_model['Sell']
X = sm.add_constant(X)
model_age = sm.OLS(y, X).fit()
print(model_age.summary())
```

### Why 
We want to understand how the age of a property affects its selling price.

### Result:
The coefficient is typically negative, meaning older homes tend to have lower selling prices.
The p-value may be higher than other predictors, suggesting this relationship may be weaker.
The R-squared is usually lower than for Taxes or Living, meaning Age alone doesn’t explain much of the variation in price.

### Conclusion:
Age has a negative but weak influence on selling price.
It’s not as good a predictor as Taxes or Living. To build a stronger model, Age should be used along with other variables.
